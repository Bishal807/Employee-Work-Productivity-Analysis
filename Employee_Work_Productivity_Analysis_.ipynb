## Data loading

### Subtask:
Load the "Enhanced_Employee_Productivity.csv" dataset into a pandas DataFrame.


**Reasoning**:
Load the dataset into a pandas DataFrame and display the first five rows.



```python
import pandas as pd

try:
    df = pd.read_csv('Enhanced_Employee_Productivity.csv')
    display(df.head())
except FileNotFoundError:
    print("Error: 'Enhanced_Employee_Productivity.csv' not found.")
    df = None
except pd.errors.ParserError:
    print("Error: Unable to parse the CSV file.")
    df = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None
```

## Data preparation

### Subtask:
Prepare the data for analysis by calculating new features.


**Reasoning**:
Calculate "Tasks_per_Hour", average "Hours_Worked", and "PEI" columns based on the given formulas, handling potential `ZeroDivisionError`.



```python
# Calculate Tasks_per_Hour
df['Tasks_per_Hour'] = df['Tasks_Completed'] / df['Hours_Worked']
df.loc[df['Hours_Worked'] == 0, 'Tasks_per_Hour'] = 0

# Calculate average hours worked
avg_hours_worked = df['Hours_Worked'].mean()

# Calculate PEI
df['PEI'] = (df['Tasks_Completed'] / df['Hours_Worked']) * df['Productivity_Score']
df.loc[df['Hours_Worked'] == 0, 'PEI'] = 0

display(df.head())
```

## Data analysis

### Subtask:
Perform exploratory data analysis on the prepared dataset.


**Reasoning**:
Calculate correlations, identify top employees by productivity score, calculate standard deviations of productivity scores by department, find the department with the minimum standard deviation, identify top employees by PEI, identify employees with high performance ratings and below average hours worked, and find the most task-efficient employee.



```python
# Calculate correlations
correlation_hours_performance = df['Hours_Worked'].corr(df['Performance_Rating'])
correlation_tasks_performance = df['Tasks_Completed'].corr(df['Performance_Rating'])
print(f"Correlation between Hours Worked and Performance Rating: {correlation_hours_performance}")
print(f"Correlation between Tasks Completed and Performance Rating: {correlation_tasks_performance}")

# Identify top 5 employees by Productivity_Score
top_employees = df.nlargest(5, 'Productivity_Score')[['Name', 'Department', 'Productivity_Score']]
print("\nTop 5 Employees by Productivity Score:")
display(top_employees)

# Calculate standard deviation of Productivity_Score for each department
std_by_department = df.groupby('Department')['Productivity_Score'].std()
print("\nStandard Deviation of Productivity Score by Department:")
print(std_by_department)

# Find the department with the minimum standard deviation
min_std_department = std_by_department.idxmin()
min_std_value = std_by_department.min()
print(f"\nDepartment with minimum standard deviation of Productivity Score: {min_std_department} ({min_std_value})")

# Identify top 3 employees based on PEI
top_pei_employees = df.nlargest(3, 'PEI')[['Name', 'Department', 'PEI']]
print("\nTop 3 Employees by PEI:")
display(top_pei_employees)

# Identify employees with Performance_Rating >= 4 and Hours_Worked < avg_hours_worked
avg_hours_worked = df['Hours_Worked'].mean()
filtered_employees = df[(df['Performance_Rating'] >= 4) & (df['Hours_Worked'] < avg_hours_worked)][['Name', 'Department', 'Hours_Worked', 'Performance_Rating']]
print("\nEmployees with Performance Rating >= 4 and Hours Worked < Average:")
display(filtered_employees)

# Find the most task-efficient employee
most_efficient = df.loc[df['Tasks_per_Hour'].idxmax()]
print("\nMost Task-Efficient Employee:")
print(f"Name: {most_efficient['Name']}, Department: {most_efficient['Department']}, Tasks_per_Hour: {most_efficient['Tasks_per_Hour']}, Performance Rating: {most_efficient['Performance_Rating']}")
```

## Data visualization

### Subtask:
Visualize the key findings from the data analysis.


**Reasoning**:
Generate the visualizations as requested in the instructions.



```python
import matplotlib.pyplot as plt

# 1. Scatter Plot (Hours Worked vs. Productivity Score)
plt.figure(figsize=(10, 6))
for department in df['Department'].unique():
    subset = df[df['Department'] == department]
    plt.scatter(subset['Hours_Worked'], subset['Productivity_Score'], label=department)
plt.xlabel('Hours Worked')
plt.ylabel('Productivity Score')
plt.title('Hours Worked vs. Productivity Score')
plt.legend()
plt.savefig('scatter_hours_productivity.png')
plt.show()

# 2. Bar Chart (Top 5 Employees by Productivity Score)
top_5_employees = df.nlargest(5, 'Productivity_Score')
plt.figure(figsize=(12, 6))
plt.bar(top_5_employees['Name'], top_5_employees['Productivity_Score'], color='skyblue')
plt.xlabel('Employee Name')
plt.ylabel('Productivity Score')
plt.title('Top 5 Employees by Productivity Score')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('bar_top_employees.png')
plt.show()

# 3. Standard Deviation Visualization (Productivity Score per Department)
std_by_department = df.groupby('Department')['Productivity_Score'].std()
plt.figure(figsize=(10, 6))
plt.bar(std_by_department.index, std_by_department.values, color='lightcoral')
plt.xlabel('Department')
plt.ylabel('Standard Deviation of Productivity Score')
plt.title('Standard Deviation of Productivity Score per Department')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('bar_std_department.png')
plt.show()

# 4. Visualization (Top 3 Employees by PEI)
top_3_pei = df.nlargest(3, 'PEI')
plt.figure(figsize=(10, 6))
plt.bar(top_3_pei['Name'], top_3_pei['PEI'], color='lightgreen')
plt.xlabel("Employee Name")
plt.ylabel("PEI")
plt.title("Top 3 Employees by PEI")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('bar_top_pei.png')
plt.show()


# 5. Visualization (Underutilized High Performers)
avg_hours = df['Hours_Worked'].mean()
underutilized = df[(df['Performance_Rating'] >= 4) & (df['Hours_Worked'] < avg_hours)]
plt.figure(figsize=(10, 6))
plt.bar(underutilized['Name'], underutilized['Hours_Worked'], color='lightgray')
plt.xlabel("Employee Name")
plt.ylabel("Hours Worked")
plt.title("Underutilized High Performers (Performance Rating >= 4)")
plt.xticks(rotation=45, ha='right')
plt.axhline(y=avg_hours, color='red', linestyle='--', label=f'Average Hours Worked ({avg_hours:.2f})')
plt.legend()
plt.tight_layout()
plt.savefig('underutilized_high_performers.png')
plt.show()

# 6. Visualization (Most Task-Efficient Employee)
most_efficient = df.loc[df['Tasks_per_Hour'].idxmax()]
plt.figure(figsize=(8, 4))
plt.bar(['Tasks per Hour'], [most_efficient['Tasks_per_Hour']], color='orange')
plt.ylabel('Tasks per Hour')
plt.title(f"Most Task-Efficient Employee: {most_efficient['Name']}")
plt.ylim(0, max(df['Tasks_per_Hour']) * 1.1)  # Adjust y-axis limit for better visualization
plt.text(0, most_efficient['Tasks_per_Hour'] , f"Department: {most_efficient['Department']}, Performance Rating: {most_efficient['Performance_Rating']}", ha='center', va='bottom')
plt.savefig('most_task_efficient.png')
plt.show()
```
